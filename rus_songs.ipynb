Подсчет метрик MTLD и TTR

```python
%pip install setuptools
```

```python
%pip install lexical_diversity
```

```python
from lexical_diversity import lex_div as ld
```

```python
lst_mtlds_ttrs = []
counter = 1
for text in texts_cr:
    tokenized_text = ld.tokenize(text)
    tokenized_text_mtld = round(ld.mtld(tokenized_text), 3)
    tokenized_text_ttr = round(ld.mattr(tokenized_text), 3)
    lst_mtlds_ttrs.append((counter, tokenized_text_mtld, tokenized_text_ttr))
    counter += 1
lst_mtlds_ttrs
```

JOIN

Я воспользовалась INNER JOIN, потому что у меня три таблицы: две с данными по песням, третья - аналитика. Я решила объединить две смежные таблицы - information и texts, присоединив id и тексты песен из второй таблицы к первой с такими данными, как id, имя автора, название, библиографические сведения и жанр.

```python
mydb = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = getpass(),
    database = 'ru_songs'
)

mycursor = mydb.cursor()
sql = "SELECT i.id, i.name, i.title, i.publication, i.year, i.g_title, t.id, t.text FROM information AS i INNER JOIN texts AS t ON i.id=t.id WHERE i.year >= 1830 ORDER BY i.year"
mycursor.execute(sql)
myresult = mycursor.fetchall()
print(myresult)
```
